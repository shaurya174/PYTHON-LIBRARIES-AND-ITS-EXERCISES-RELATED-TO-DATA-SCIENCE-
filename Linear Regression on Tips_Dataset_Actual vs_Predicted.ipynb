```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Model Prerequisites From Sckit Learn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
#Data
df = sns.load_dataset('tips')
```

```python
df
```

```python
#X values
x = df[['total_bill','size']].values
```

```python
x
```

```python
# Y values
y = df['tip'].values
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train
```

```python
y_train
```

```python
y_test
```

```python
x_test
```

```python
model.fit(x_train,y_train)
```

```python
outputs=model.predict(x_test)
```

```python
outputs
```

```python
indexes = np.arange(1,75,1)
indexes
```

```python
#Compare Outputs and Actual Values
%matplotlib inline
plt.scatter(indexes, y_test, color='blue', marker='o', label='Actual Values')
plt.scatter(indexes, outputs, color='red', marker='x', label='Predicted Values')
plt.xlabel('Indexes')
plt.ylabel('Values')
plt.title('Comparing Actual And Predicted Values')
plt.legend()
plt.grid(True)
plt.show()
```
