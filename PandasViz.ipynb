```python
#Import Numpy And Pandas
import numpy as np
import pandas as pd
```

```python
#Load df3 
df3 = pd.read_csv('df3')
```

```python
#Import matplotlib
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#Check head of df3
df3.head()
```

```python
#Scatter plot of b vs a (COLOR AND SIZE CHANGED) AND USE FIGSIZE TO STRECH IT
df3.plot.scatter(x='a',y='b',c='red',s=40,figsize=(12,2))
```

```python
#Histogram Of 'a' Column
df3['a'].hist(bins=30)
```

```python
#BoxPlot Of 'a' and 'b' columns
df3[['a','b']].plot(kind='box')
```

```python
#kde plot of 'd' column
df3['d'].plot.kde()
```

```python
#Area plot of all the columns for just the rows up to 30
df3.head(30).plot.area()
```
