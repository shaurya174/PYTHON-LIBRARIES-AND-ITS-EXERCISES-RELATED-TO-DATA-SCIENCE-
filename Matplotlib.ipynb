#MATPLOTLIB NOTEBOOK WITH SOME BASIC CURVES

Imp -> ALL THE COMMANDS FOR PLOTTING A FIGURE SHOULD ALL GO IN THE SAME CELL. SEPARATING THEM OUT INTO MULTIPLE CELLS MAY CAUSE NOTHING TO SHOW UP

```python
#Sample Data To Work With
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
#Import Matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Follow along with these steps: **
#1. Create a figure object called fig using plt.figure()
#2. Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax
#3. Plot (x,y) on that axes and set the labels and titles
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
#Figure -> 2
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('zoom')
ax1.plot(x,z)
ax2.plot(x,y)
```

```python
#Figure -> 4
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,color='red',ls='--')
axes[1].plot(x,z,color='purple',ls=':')
fig.tight_layout()
```

```python
import numpy as np
```

```python
#Sine Wave Values
x = np.linspace(-1,1,100)
y = np.sin(x)
print(x)
print(y)
```

```python
#Cosine Wave Values
a = np.linspace(-1,1,100)
b = np.cos(x)
print(a)
print(b)
```

```python
#Tangent Wave Values
c = np.linspace(-1,1,100)
d = np.tan(x)
print(c)
print(d)
```

```python
#Exponent Wave Values
g = np.linspace(-1,1,100)
h = np.exp(g)
print(g)
print(h)
```

```python
#FINAL CURVE
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y,color='purple',linestyle='--',label='Sin Wave')
axes[0][0].legend(loc=0)
axes[0][1].plot(a,b,color='red',linestyle='-.',label='Cosine Wave')
axes[0][1].legend(loc=0)
axes[1][0].plot(c,d,color='magenta',linestyle=':',label='Tangent Wave')
axes[1][0].legend(loc=0)
axes[1][1].plot(g,h,color='teal',linestyle='-',label='e^x Wave')
axes[1][1].legend(loc=0)
fig.tight_layout()
```
